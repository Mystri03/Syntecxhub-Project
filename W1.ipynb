## Step 6: Data Filtering

```python
high_spenders = data[data["Purchase Amount (USD)"] > 80]
high_spenders.head()
```

```python
female_clothing = data[(data["Gender"] == "Female") & (data["Category"] == "Clothing")]
female_clothing.head()
```

## Step 7: Pivot Table & GroupBy Analysis

```python
data.groupby("Gender")["Purchase Amount (USD)"].mean()
```

```python
data.groupby("Category")["Purchase Amount (USD)"].sum()
```

```python
pd.pivot_table(data,values='Payment Method',index='Gender',aggfunc=lambda x: x.mode()[0])
```

```python
pd.pivot_table(data,values='Gender',index='Season',aggfunc='count')
```

## Step 8: Data Sorting

## Step 12: Standardize Column Names

```python
data.columns = data.columns.str.strip()
data.columns = data.columns.str.lower()
data.columns = data.columns.str.replace(" ", "_")
data.columns = data.columns.str.replace("(", "")
data.columns = data.columns.str.replace(")", "")
```

```python
# Convert column names to lowercase and replace spaces with underscores
data.columns = data.columns.str.lower().str.replace(" ","_")
data.columns
```

## Step 13: Simple Visualization

### 1. Histogram
- Shows distribution of a numeric column.

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8,3))
plt.hist(data['age'], bins=30)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

### 2. Bar Chart
- Count of Gender

```python
data['gender'].value_counts().plot(kind='bar')
plt.title('Gender Count')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

### 3. Pie Chart 
- Distribution of Payment Methods

```python
plt.figure(figsize=(8,3))
data['payment_method'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Payment Method Distribution')
plt.ylabel('') 
plt.show()
```
